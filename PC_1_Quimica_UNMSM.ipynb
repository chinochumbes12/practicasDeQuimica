```python
# Guarda como variable la masa atómica de los elementos C, H y O, e imprime la masa molecular del etanol.
# Masas atómicas de los elementos
masa_atomica_C = 12.01  # u
masa_atomica_H = 1.01   # u
masa_atomica_O = 16.00  # u

# Fórmula química del etanol (C2H5OH)
num_carbonos = 2
num_hidrogenos = 6
num_oxigenos = 1

# Calcular la masa molecular del etanol
masa_molecular_etanol = (num_carbonos * masa_atomica_C) + (num_hidrogenos * masa_atomica_H) + (num_oxigenos * masa_atomica_O)

# Imprimir la masa molecular del etanol
print("Masa molecular del etanol (C2H5OH):", masa_molecular_etanol, "u")
```

```python
# Busca la temperatura de fusión en Celsius de 5 metales y guárdalos en una lista. Imprime las temperaturas en Kelvin
# Lista de metales y sus temperaturas de fusión en grados Celsius
metales_temperaturas = [
    ("Hierro", 1538),
    ("Cobre", 1084),
    ("Aluminio", 660.32),
    ("Plata", 961.78),
    ("Oro", 1064)
]

# Función para convertir temperaturas de Celsius a Kelvin
def celsius_a_kelvin(celsius):
    return celsius + 273.15

# Imprimir las temperaturas en Kelvin junto con los nombres de los metales
for metal, temp_celsius in metales_temperaturas:
    temp_kelvin = celsius_a_kelvin(temp_celsius)
    print(f"{metal}: {temp_kelvin:.2f} K")
```

## Pregunta 2: Operaciones (4 puntos)

Escribe un programa que lea repetidamente elementos químicos en español hasta que el usuario introduzca “fin”. Una vez se haya introducido “fin”, muestra por pantalla (o sea imprime) el total de elementos ingresados. Si el usuario introduce cualquier otra cosa que no sea un elemento químico, detecta su fallo usando try y except, muestra un mensaje de error y pasa al siguiente elemento.

```python
# Diccionario de elementos químicos en español
elementos_quimicos = {
    "hidrogeno": "H",
    "helio": "He",
    "litio": "Li",
    "berilio": "Be",
    "boro": "B",
    "carbono": "C",
    "nitrogeno": "N",
    "oxigeno": "O",
    "flúor": "F",
    "neon": "Ne",
    "sodio": "Na",
    "magnesio": "Mg",
    "aluminio": "Al",
    "silicio": "Si",
    "fosforo": "P",
    "azufre": "S",
    "cloro": "Cl",
    "argon": "Ar",
    "potasio": "K",
    "calcio": "Ca",
    # Puedes agregar más elementos aquí
}

# Variable para almacenar el contador de elementos ingresados
total_elementos = 0

while True:
    # Solicitar al usuario que ingrese un elemento químico en español
    elemento = input("Ingrese un elemento químico en español (o 'fin' para finalizar): ").lower()

    if elemento == 'fin':
        break  # Salir del bucle si se ingresa "fin"

    try:
        # Intentar obtener el símbolo químico del diccionario
        simbolo = elementos_quimicos.get(elemento)
        if simbolo is not None:
            print(f"Símbolo químico de {elemento}: {simbolo}")
            total_elementos += 1
        else:
            raise KeyError()  # Lanzar una excepción si el elemento no está en el diccionario
    except KeyError:
        # Manejar el error si el elemento no se encuentra en el diccionario
        print(f"Error: '{elemento}' no es un elemento químico válido.")

# Mostrar el total de elementos ingresados
print(f"Total de elementos químicos ingresados: {total_elementos}")
```
