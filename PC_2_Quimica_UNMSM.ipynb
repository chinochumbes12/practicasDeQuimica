```python
from rdkit import Chem

# Obtener la tabla periódica
PeriodicTable = Chem.GetPeriodicTable()

# Crear una lista para almacenar los símbolos de los elementos
element_symbols = []

# Obtener los símbolos de los primeros 20 elementos
for i in range(1, 21):
    element_symbols.append(PeriodicTable.GetElementSymbol(i))

# Imprimir la lista generada
print(element_symbols)
```

```python
from sympy import symbols, Eq, solve

# Definir los símbolos de los coeficientes
x, y, z = symbols('x y z')

# Definir la ecuación de balanceo de la masa
eq1 = Eq(3*x, 8)  # Balanceo de los átomos de oxígeno
eq2 = Eq(2*y, 2)  # Balanceo de los átomos de carbono
eq3 = Eq(2*z, 8)  # Balanceo de los átomos de hidrógeno

# Resolver el sistema de ecuaciones
sol = solve((eq1, eq2, eq3), (x, y, z))

# Obtener los coeficientes balanceados
coef_O2 = sol[x]
coef_CC = sol[y]
coef_O = sol[z]

# Imprimir la ecuación balanceada
print(f"{coef_O2} O2 + {coef_CC} CC=CC + {coef_O} O=O → 8C(=O)=O + 8H2O + 9O")
```
